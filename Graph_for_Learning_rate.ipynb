```python
#import modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load data
data1 = np.load('bin_LV_098.npy', allow_pickle=True)
data2 = np.load('bin_RV_098.npy', allow_pickle=True)
```

```python
#calculate mean,std and 95% confidence interval
mean1 = np.mean(data1, axis=0)
std1 = np.std(data1, axis=0)
confidence_interval1 = 1.96 * std1 / np.sqrt(50)

mean2 = np.mean(data2, axis=0)
std2 = np.std(data2, axis=0)
confidence_interval2 = 1.96 * std2 / np.sqrt(50)

```

```python
#sort x-axis
x = np.arange(1,51)

#plot means
plt.plot(x,mean1)
plt.plot(x,mean2)

#plot confidence
plt.fill_between(x, mean1 - confidence_interval1, mean1 + confidence_interval1,alpha=0.2,label='Lane Change Vehicle')
plt.fill_between(x, mean2 - confidence_interval2, mean2 + confidence_interval2,alpha=0.2,label='Give Way Vehicle')

#title and lavels
plt.title('Bins starting at 0.098 and 0.02')
plt.xlabel('Episode number')
plt.ylabel('Average Total Episode Reward')

#legend and make it start at 0
plt.xlim(left=0)
plt.legend()

#save or show
#plt.show()
plt.savefig('bins_098.pdf')
```
